```python
import pandas as pd
from transformers import BartTokenizer
```

```python
df = pd.DataFrame(data)
df = df.set_index('Unnamed: 0')
```

```python
df.head()
```

```python
# BART 모델의 토크나이저 로드
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large')

# text 열을 토큰화
df['tokenized_text'] = df['text'].apply(lambda x: tokenizer.encode(x, truncation=True, padding='max_length'))

# 토큰화된 텍스트 확인
print(df[['text', 'tokenized_text']])
```

```python
# 마찬가지로 title도 토큰화
df['tokenized_title'] = df['title'].apply(lambda x: tokenizer.encode(x, truncation=True, padding='max_length'))
print(df[['title', 'tokenized_title']])
```
