```python
# Unnamed:0을 인덱스로
df = df.set_index('Unnamed: 0')
```

```python
# 결측치 여부
null_title = df['title'].isnull().sum()
null_text = df['text'].isnull().sum()

print(null_title)
print(null_text)
```
